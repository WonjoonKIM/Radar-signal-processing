```python
# 레이다 방사 전력밀도 계산
import numpy as np
import math as mt

# 2. 안테나로부터 근접 전계 구역 간의 거리_X(m)_책657 (1)과 (2)사이
# R은 전력밀도를 계산하는 지점 간의 거리 (m)
# D는 안테나 길이 (m)
# L(λ)는 파장 (m), 5Ghz는 파장 10 x 3/5 = 6cm
R = 1
D = 0.24
L = 0.06
def findX(R,D,L):
    X=R/(2*np.power(D,2)/L)
    return X

X=findX(R, D, L)
print("안테나로부터 근접 전계 구역 간의 거리 = ", X,"m")

# 1. 근접 전계 구간에서 전력밀도와 최대 전력밀도 계산_책657 (1), 책 657(2)
P0=(16/X)*mt.sin(np.pi/8*X)
P1=1-P0
P2=(128*np.power(X,2))/np.power(np.pi,2) 
P3=1-(mt.cos(np.pi/8*X))
P4=P1+(P2*P3)
P=P4*26.1
print("근접 전계 구간에서 전력밀도 = ", P,"(W/M^2)")
print(P0,P1,P2,P3,P4)

P_RFmax = (0.2*np.power(D,2))/L
print("최대 전력밀도 계산 = ",P_RFmax,"(W/M^2)")

# 3. 원 전계에서 전력밀도_책658 (3)
# Pavg는 평균전력
# Gt는 레이다 송신 안테나의 이득
# R은 송신 안테나로부터 거리  (m)
Pavg= 10
Gt=15 #15db
R= 2
Pff=(Pavg*Gt)/(4*np.pi*np.power(R,2))
print("원 전계에서 전력밀도 = ", Pff,"(W/M^2)")

# 4. 원 전계 경계지점과 원 전계 경계지점에서 전력밀도_책658 (3)과 (4)사이, 책658 (4)
far_field_R=(2*np.power(D,2))/L
Pffb=(Pavg*Gt*np.power(L,2))/(16*np.pi*np.power(D,2))
print("원 전계 경계지점 = ", far_field_R, "M")
print("원 전계 경계지점에서 전력밀도 = ", Pffb,"(W/M^2)")

# 5. 근접 전계 구간 거리에서 전력밀도_책658 (5)
Pnf=Pffb*P
print("근접 전계 구간 거리에서 전력밀도 = ", Pnf,"(W/M^2)")
# 원 전계 경계 지점의 전력밀도(4)와 근접 전계 구간의 거리에 따른 가중치(2)를 곱하여 구할 수 있다.

# 6. 수신 전력의 크기_워드 
#Pt : 첨두 송신 출력
#Ae : 안테나 개구 효율 (m2)
#λ : 파장
#G : 안테나 이득
#σ(시그마) : 표적의 반사 단면적(RCS)
#r : 타켓과의 거리 (m)
Pt = 44 #여기서는 최대 전력밀도를 입력하였다.
G = 15 # 혼 안테나의 경우
sigma = 1 # 사람의 경우 1
r= 1

Pradar_wo_Ae_upper=(Pt*G*sigma)
Pradar_wo_Ae_lower=np.power((4*np.pi*np.power(r,2)),2)
Pradar_wo_Ae=Pradar_wo_Ae_upper/Pradar_wo_Ae_lower

Ae=(np.power(L,2)*G)/(4*np.pi)
Pradar_wt_Ae=Pradar_wo_Ae*Ae
print("수신 전력의 크기 = ",round(Pradar_wt_Ae,4),"W")
#print(Pradar_wo_Ae_upper, Pradar_wo_Ae_lower, Pradar_wo_Ae, Ae)
second_Pradar=(Pt*np.power(G,2)*np.power(L,2)*sigma)/(np.power(4*np.pi, 3)*np.power(r,4))
#print(round(second_Pradar,10)) #0.0000816367W

# 7. 최소 탐지 신호 파워_워드
Pmin_upper=Pt*np.power(G,2)*np.power(L,2)*sigma
Pmin_lower=(np.power(4*np.pi,3))*(np.power(r,4))
Pmin=Pmin_upper/Pmin_lower
print("최소 탐지 신호 파워 = ", round(Pmin,4),"W")
#print(Pmin_upper, Pmin_lower, Pmin)

# 8. 최대 탐지 거리_워드
Rmax_upper=Pmin_upper
Rmax_lower=(np.power(4*np.pi,3))*(Pmin)
Rmax=np.power(Rmax_upper/Rmax_lower,0.25)
print("최대 탐지 거리 = ", Rmax,"m")
```

```python
import numpy as np
import math as mt

#Non-directional Power Density Su
Ps=0.1 # SDR 송신 전력이 최대 100mw
R=0.5 #1m에서 측정한다고 가정
Su=Ps/(4*mt.pi*mt.pow(R,2))
#print("무지향성 파워 밀도 su : ", Su)

#the directional power density Sg
G=15 #혼 안테나는 15db
Sg=Su*G
print("지향성 파워 밀도 sg : ",round(Sg,3))

# the reflected power Pr 
RCS=1 #사람의 경우 1
Pr=(Ps*G*RCS)/(4*mt.pi*mt.pow(R,2))
print("반사되는 파워 pr : ", round(Pr,3))

#power density at receiving place
Se=Pr/(4*mt.pi*mt.pow(R,2))
print("수신되는 파워 se : ", round(Se,3))

#power density at the receiving site Pe
#AW :  effective antenna aperture [m²]
AW=0.7
Pe=Se*AW
print("수신 안테나에서 측정되는 파워 Pe : ", round(Pe,3))

#최대 탐지거리
L=0.06
R=mt.sqrt((Ps*mt.pow(G,2)*mt.pow(L,2)*RCS)/(Pe*mt.pow(4*mt.pi,3)))/4
#R=mt.sqrt((Ps*mt.pow(G,2)*mt.pow(L,2)*RCS)/(Pe*mt.pow(4*mt.pi,3)))/4
print("최대 탐지 거리 : ", round(R,4))

#upper=Ps*mt.pow(G,2)*mt.pow(L,2)*RCS
#lower=Pe*mt.pow(4*mt.pi,3)
#print(upper,lower)

rmax=(0.1*mt.pow(G,2)*mt.pow(L,2)*1)/mt.pow((4*mt.pi),2)*0.018
Rmax=rmax/4
print(round(Rmax,4))
```
